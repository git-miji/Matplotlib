# Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible. It is a low level graph plotting library in python that serves as a visualization utility. Matplotlib is open source and we can use it freely.

Matplotlib was created by John D. Hunter.

## Pyplot

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

import matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(1,20)
y=np.arange(61,80)
print(x)
print(y)
```

```python
a=np.arange(11,30)
b=np.arange(21,40)
print(a)
print(b)
```

```python
# Plotting Using Matplotlib

# Plt Scatter

plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2-D')
plt.savefig('test.jpg')
```

```python
z=x*x
w=x*y
print(z)
print(w)
plt.plot(w,z)
```

# Creating Subplot

```python
plt.subplot(2,2,1)
plt.plot(w,x,'b')
plt.subplot(2,2,2)
plt.plot(z,y,'g')
plt.subplot(2,2,3)
plt.plot(z,x,'r--')
plt.subplot(2,2,4)
plt.plot(x,y,'r--')
```

```python
# Compute the x and y coordinates for points on sine wave
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave')
plt.plot(x,y)
plt.show()
```

```python
from numpy import random
from mpl_toolkits import mplot3d
```

```python
ax=plt.axes(projection='3d')
x=np.arange(1,10,0.1)
y=np.arange(11,20,0.1)
z=np.arange(21,30,0.1)
ax.plot(x,y,z)
plt.show()
```

```python
#Using Random function
ax=plt.axes(projection='3d')
x=np.random.randint(1,10,(300,))
y=np.random.randint(1,10,(300,))
z=np.random.randint(1,10,(300,))
p=np.random.randint(11,20,(300,))
q=np.random.randint(11,20,(300,))
r=np.random.randint(11,20,(300,))

ax.scatter(x,y,z,p,q,r, cmap="plasma")
plt.show()
```

# Ploting a Surface

```python
ax=plt.axes(projection='3d')
x=np.arange(1,10)
y=np.arange(11,20)
X,Y=np.meshgrid(x,y)
Z=X*Y
ax.plot_surface(X,Y,Z)
plt.show()
```
